2D Kinematics Equations for parabolic motion
    y(t) =  y_0 + v_y_0 + g/2 ** t
    x(t) = x_0 + v_x_0 * t

```python
import matplotlib.pyplot as plot
import numpy as np

v_x_0 = 10 #m
v_y_0 = 10 #m
v_x = 0 #m/s
v_y = 0 #m/s
g = (9.8) #m/s^-2

times = np.linspace(0, 2, 30)
x_values = []
y_values = []

for time in times:
    x_values.append(v_x + v_x_0 * time)
    y_values.append(v_y + v_y_0 * time - (g/2 * time ** 2))

plot.plot(x_values, y_values)
plot.title("Projectile Motion Graph")
plot.xlabel("x/m")
plot.ylabel("y/m")

plot.show()
```
