```python
import numpy as np

#gravity
g = (9.8) #m/s^-2

#mass, velocity, height
m = float(input("How heavy is the object? ")) #kg
v = float(input("How fast is the object intially going? ")) #m/s
h = float(input("What height is the object located initially? ")) #m

#Kinetic Energy
KE = (1/2) * m * (v**2) #Joules

#Gravitational Potential Energy
UE = m * g * h


#returns total energy
def total_energy():
    return KE + UE

print(total_energy())

```
