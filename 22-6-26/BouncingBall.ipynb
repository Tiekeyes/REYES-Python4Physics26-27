```python
import numpy as np
import matplotlib.pyplot as plot
import matplotlib.animation as animation


class PhysicsObject:

    def __init__(self, mass, elasticity, initial_velocity=(0, 0), initial_pos=(0, 0)):
        self.mass = mass #kgs
        self.position = list(initial_pos) #m
        self.gravity = 9.8 #m/s^-2
        self.v_x = initial_velocity[0] #m/s
        self.v_y = initial_velocity[1] #m/s
        self.elas = elasticity
    

    #delta time is in seconds
    def update(self, delta_time=1):
        #find x position
        self.position[0] += self.v_x * delta_time

        #hit or passed the ground
        if(self.position[0] <= 0):
            self.v_y *= self.v_y * self.elasticity * -1 
        
        #find y position
        self.position[1] += (self.v_y * delta_time) - (0.5 * self.gravity * delta_time**2)

        self.v_y -= self.gravity * delta_time


ball = PhysicsObject(5, 0.3, (1, 1))
while(ball.position[1] > -10):
    ball.update(0.01)
    print(ball.position)
    
```
