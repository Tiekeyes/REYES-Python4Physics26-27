Now, create a dictionary to organize the sports into similar categories (i.e. skating sports, ball sports, etc.)

Recall dictionary structure: my_dictionary = {"Key": "Value", "Key2": "Value 2" "Value 3"}

```python
sports_dict = {"Ball": ["Soccer", "Tennis"], "Running": ["Cross Country", "Track"]}

print(sports_dict["Running"][1])
my_sport_type = list(sports_dict.keys())[1]
print(my_sport_type == "Running")

def id_type_of_key(target, dict):
    
    for key in dict.keys():
        if target in dict[key]:
            return key
    return None

print(id_type_of_key("Tennis", sports_dict))
```
